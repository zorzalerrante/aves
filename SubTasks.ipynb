```python
#Funcion para asignar tipo de día
def etiquetar_tipo_dia(row):
    tipos_dia = []
    if pd.notna(row['FactorLaboralNormal']):
        tipos_dia.append('Laboral')
    if pd.notna(row['FactorDomingoNormal']):
        tipos_dia.append('Domingo')
    if pd.notna(row['FactorSabadoNormal']):
        tipos_dia.append('Sábado')
    if pd.notna(row['FactorLaboralEstival']):
        tipos_dia.append('LaboralEstival')
    if pd.notna(row['FactorFindesemanaEstival']):
        tipos_dia.append('FindesemanaEstival')
    if not tipos_dia:
        tipos_dia.append('No Definido')
    return tipos_dia

# Aplicamos la función a cada fila y creamos una nueva columna llamada 'TipoDia'
viajes_stages['TipoDia'] = viajes_stages.apply(etiquetar_tipo_dia, axis=1)

viajes_stages[['Viaje', 'Etapas', 'Etapa', 'Periodo', 'TipoDia']].sample(5)
```

```python
#Como no hay tipos de dias con mas de una etiqueta parece más apropiado usar 

#Funcion para asignar tipo de día
def etiquetar_tipo_dia(row):
    if pd.notna(row['FactorLaboralNormal']):
        return 'Laboral'
    if pd.notna(row['FactorDomingoNormal']):
        return 'Domingo'
    if pd.notna(row['FactorSabadoNormal']):
        return 'Sábado'
    if pd.notna(row['FactorLaboralEstival']):
        return 'LaboralEstival'
    if pd.notna(row['FactorFindesemanaEstival']):
        return 'FindesemanaEstival'
    else:
        return 'No Definido'

def etiquetar_FactorExt(row):
    if pd.notna(row['FactorLaboralNormal']):
        return row['FactorLaboralNormal']
    if pd.notna(row['FactorDomingoNormal']):
        return row['FactorDomingoNormal']
    if pd.notna(row['FactorSabadoNormal']):
        return row['FactorSabadoNormal']
    if pd.notna(row['FactorLaboralEstival']):
        return row['FactorLaboralEstival']
    if pd.notna(row['FactorFindesemanaEstival']):
        return row['FactorFindesemanaEstival']
    else:
        return None

# Aplicamos la función a cada fila y creamos una nueva columna llamada 'TipoDia'
viajes_stages['TipoDia'] = viajes_stages.apply(etiquetar_tipo_dia, axis=1)

viajes_stages['FactorExterno'] = viajes_stages.apply(etiquetar_FactorExt, axis=1)

viajes_stages[['Viaje', 'Etapas', 'Etapa', 'Periodo', 'TipoDia', 'FactorExterno']].sample(5)
```
